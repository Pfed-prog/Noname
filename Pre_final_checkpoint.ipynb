```python
import pandas as pd
df0 = pd.read_excel('df12.xlsx', sheet_name=0)
df1 = pd.read_excel('df12.xlsx', sheet_name=1)
df2 = pd.read_excel('df12.xlsx', sheet_name=2)
df3 = pd.read_excel('df12.xlsx', sheet_name=3)
```

```python
df0.columns=['Id посылки', 'Id офиса отправления', 'Дата отправления',
       'id офиса получения', 'Дата получения', 'Время в пути']
df1.columns=['Id посылки', 'Тип посылки', 'ID начального офиса',
       'ID конечного офиса', 'Озвученная дата получения']

df0['Дата отправления'] = df0['Дата отправления']+ pd.DateOffset(hours=0.1)
df0['Дата получения'] = df0['Дата получения']+ pd.DateOffset(hours=0.1)
df1['Озвученная дата получения'] = df1['Озвученная дата получения']+ pd.DateOffset(hours=0.1)

df3['Time calculation request'] = df3['Time calculation request']+ pd.DateOffset(hours=0.1)
df3['Time delivered to office'] = df3['Time delivered to office']+ pd.DateOffset(hours=0.1)
```
